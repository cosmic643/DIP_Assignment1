**Scaling Up using Replication**

```python
import cv2
import numpy as np
img = cv2.imread("lena_color.tif",0)
if img is None:
    print("Image Loading Failed")
    exit(0)
height,width = img.shape
new_img = np.zeros((2*height,2*width))
#img.shape[0] = height and img.shape[1] = width
for i in range(height):
    for j in range(width):
        new_img[2*i][2*j] = img[i][j]
        new_img[2*i + 1][2*j] = img[i][j]
        new_img[2*i][2*j + 1] = img[i][j]
        new_img[2*i + 1][2*j + 1] = img[i][j]
cv2_imshow(img)
cv2_imshow(new_img)
```
